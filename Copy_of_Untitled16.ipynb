i will handle missing valuse by imputation

```python
median_age = df ['Age'].median()
df['Age'].median()
```

```python
df['Age'] = df['Age'].fillna(median_age)
```

```python
mode_cabin = df['Cabin'].mode()[0]
print(f"The mode of the 'Cabin' column is: {mode_cabin}")
```

for imputation for the cabin, the median cannot be used because its a string value so the median cannot be determined rather the mode will be used

```python
df['Cabin']=df['Cabin'].fillna(mode_cabin)
```

```python
df.isnull().sum()
```

```python
mode_Embarked = df['Embarked'].mode()[0]
print(f"The mode of the 'Embarked' column is: {mode_Embarked}")
```

```python
df['Embarked']=df['Embarked'].fillna(mode_Embarked)
```
